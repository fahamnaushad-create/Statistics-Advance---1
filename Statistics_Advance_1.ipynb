##1. What is a random variable in probability theory?

- A random variable is a numerical quantity that represents the outcome of a random process or experiment. It assigns a real number to each possible outcome, allowing probability theory to work with numerical values. Random variables can be used to compute probabilities, expectations, and variances. They form the foundation for probability distributions and statistical inference.

##2. What are the types of random variables?

- There are two main types: discrete and continuous. Discrete random variables take countable values (like number of heads in coin flips). Continuous random variables take values from a range or interval (like height or weight) and are described by probability density functions. Some contexts treat mixed types too, where a variable can have both discrete and continuous components.

##3. Explain the difference between discrete and continuous distributions.

- Discrete distributions assign probabilities to distinct, separate outcomes (e.g., Poisson or binomial). Probabilities are sums of point probabilities. Continuous distributions assign probabilities over intervals using a probability density function (e.g., normal distribution), and probabilities are computed as integrals. Choice depends on whether outcomes are countable or measured on a continuum.

##5. What is the standard normal distribution, and why is it important?

- The standard normal distribution is a normal distribution with mean 0 and standard deviation 1. It is central because many statistical methods and tables are built on it; other normal distributions can be converted to it via standardization (z-scores). It appears in the Central Limit Theorem and is widely used for hypothesis tests and confidence intervals.

##6. What is the Central Limit Theorem (CLT), and why is it critical in statistics?

- The CLT states that the distribution of the sample mean approaches a normal distribution as sample size increases, regardless of the parent distribution‚Äôs shape (provided variance is finite). It allows us to use normal-based inference for sample means and construct confidence intervals, even when the underlying data are not normal, as long as sample size is reasonably large.

##7. What is the significance of confidence intervals in statistical analysis?

- A confidence interval gives a range of plausible values for an unknown population parameter (like a mean) with a specified confidence level (e.g., 95%). It quantifies uncertainty in estimates and helps decision-makers understand how precise an estimate is. Confidence intervals are more informative than point estimates because they show both estimate and uncertainty.

##8. What is the concept of expected value in a probability distribution?

- The expected value (or mean) of a random variable is the long-run average if the experiment is repeated many times. It‚Äôs computed as a weighted average of all possible values, with probabilities as weights. Expected value summarizes the central tendency of a distribution and is used in decision theory, risk analysis, and many statistical calculations.

##9. Write a Python program to generate 1000 random numbers from a normal distribution with mean = 50 and standard deviation = 5. Compute its mean and standard deviation using NumPy, and draw a histogram to visualize the distribution.

- Python code

```python
import numpy as np
import matplotlib.pyplot as plt

# Q9: generate 1000 samples from N(50, 5)
np.random.seed(0)  # reproducible
samples = np.random.normal(loc=50, scale=5, size=1000)

mean_sample = samples.mean()
std_sample = samples.std(ddof=1)  # sample standard deviation

print("Q9: 1000 samples from N(mean=50, sd=5)")
print(f"Computed mean: {mean_sample:.4f}")
print(f"Computed sample std (ddof=1): {std_sample:.4f}")

# Plot histogram
plt.figure(figsize=(8,4))
plt.hist(samples, bins=30, edgecolor='black')
plt.title("Histogram of 1000 samples from N(50, 5)")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```

##10. You are working as a data analyst for a retail company. The company has collected daily sales data for 2 years and wants you to identify the overall sales trend.
 -daily_sales = [220, 245, 210, 265, 230, 250, 260, 275, 240, 255,
235, 260, 245, 250, 225, 270, 265, 255, 250, 260]

  -  Explain how you would apply the Central Limit Theorem to estimate the average sales
with a 95% confidence interval.
  
  - Write the Python code to compute the mean sales and its confidence interval.

  ###- To estimate the average daily sales and quantify uncertainty, you compute the sample mean and sample standard error (sample std divided by sqrt(n)). By CLT, the sampling distribution of the mean is approximately normal for moderate/large n. For small samples (like n=20 here) we use the t-distribution with ùëë ùëì = ùëõ ‚àí 1 to get the critical value for a 95% confidence interval. The interval is mean ¬± (t_crit * standard_error). This gives the range where the population mean likely lies with 95% confidence.

‚úÖPython code

```python
import numpy as np
from math import sqrt

daily_sales = [220, 245, 210, 265, 230, 250, 260, 275, 240, 255, 235, 260, 245, 250, 225, 270, 265, 255, 250, 260]

arr = np.array(daily_sales)
n = len(arr)
sample_mean = arr.mean()
sample_std = arr.std(ddof=1)
se = sample_std / sqrt(n)

# Use t critical value when possible; fall back to z approx if scipy missing
try:
    from scipy import stats
    t_crit = stats.t.ppf(0.975, df=n-1)
    note = f"Used t critical value with df={n-1}."
except Exception:
    t_crit = 1.96
    note = "scipy not available: used z=1.96 as approximation."

lower = sample_mean - t_crit * se
upper = sample_mean + t_crit * se

print("Q10 results:")
print(f"n = {n}")
print(f"Sample mean = {sample_mean:.4f}")
print(f"Sample std (ddof=1) = {sample_std:.4f}")
print(f"Standard error = {se:.4f}")
print(note)
print(f"95% Confidence Interval for the mean: [{lower:.4f}, {upper:.4f}]")
```
